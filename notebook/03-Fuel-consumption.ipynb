```python
import os
from scipy.io import loadmat

import pandas as pd 
import numpy as np 
import dask.array as da 
import dask.dataframe as dd

from datetime import datetime, timedelta

# Plots 
import matplotlib.pyplot as plt 
from matplotlib import rc 

rc('font',**{'family':'serif','serif':['Times']})
rc('text', usetex=True)

from holoviews.operation import gridmatrix
```

Loading all data

```python
data_dir_list = ('..','raw','20160316_061540_DE477VE_Description.mat')
kernel_path = os.getcwd()
data_dir_path = os.path.join(kernel_path,*data_dir_list)
dct_data = loadmat(data_dir_path, matlab_compatible= True, squeeze_me = True)
```

List of all posible variables

```python
varname_lst = []    
for fld in dct_data['DAY'].dtype.fields:
    varname_lst.append(fld)
```

Extracting all position data

```python
gps_var = ['TIME','DATE_HOUR_GPS', 'SPEED', 'FUEL_FLOW','ALTITUDE','SLOPE']
lst_gps = [data.transpose()[0].transpose() for var, data in zip(varname_lst, dct_data['DAY'][0][0]) if var in gps_var]
```

Cleaning the list of values. Droping out all `nan`

```python
gps_flt = [row for row in zip(*lst_gps) if not np.isnan(row).any()]
```

Transform into data frame 

```python
gps_df = pd.DataFrame(gps_flt, columns = gps_var)
```

```python
gps_df = gps_df.drop_duplicates()
```

Transform the date 

```python
def transform_date(matlab_datenum):
    """ Convert a date to datetime """
    python_datetime = datetime.fromordinal(int(matlab_datenum)) + timedelta(days=matlab_datenum%1) - timedelta(days = 366)
    return python_datetime

gps_df['TIME_CNV'] = gps_df['DATE_HOUR_GPS'].apply(transform_date)
gps_df_flt = gps_df.set_index('TIME_CNV')
```

Compute acceleration

```python
def compute_acceleration(df):
    """ Compute acceleration"""
    num = (df['SPEED']- df['SPEED'].shift(1))/ 3.6 # Delta x
    den = (df['TIME']- df['TIME'].shift(1)) # Delta t
    df = (num).divide(den)
    return df
```

```python
gps_df_flt['ACCELERATION'] = compute_acceleration(gps_df_flt)
```

Filtering data according to a time interval for fitting analysis

```python
def transform_timestamp(matlab_datenum):
    """ Convert a date to datetime string """
    python_datetime = datetime.fromordinal(int(matlab_datenum)) + timedelta(days=matlab_datenum%1) - timedelta(days = 366)
    return python_datetime.strftime('%H:%M:%S')


plt.figure(figsize = (15,5))
# Speed 
plt.subplot(141)
ax1 = plt.gca()
st_time = '2016-03-16 16:00:00'
ed_time = '2016-03-16 18:40:00'
gps_df_flt.loc[st_time:ed_time].plot(y='SPEED',ax=ax1);
locs, labels = plt.xticks();
plt.xticks(locs,map(transform_timestamp,locs));
plt.xlabel(r'Time [HH:MM]',fontsize=14);
plt.ylabel(r'Speed [$Km/h$]',fontsize=14);
plt.legend([r'$v_k$'], fontsize=14);
plt.grid(True, linestyle='--')

# Acceleration
plt.subplot(142)
ax1 = plt.gca()
gps_df_flt.loc[st_time:ed_time].plot(y='ACCELERATION',ax = ax1);
locs, labels = plt.xticks();
plt.xticks(locs,map(transform_timestamp,locs));
plt.xlabel(r'Time [HH:MM]',fontsize=14);
plt.ylabel(r'Acceleration [$km/h^2$]',fontsize=14);
plt.legend([r'$a_k$'], fontsize=14);
plt.grid(True, linestyle='--')

# Slope 
plt.subplot(143)
ax1 = plt.gca()
gps_df_flt.loc[st_time:ed_time].plot(y='SLOPE',ax=ax1);
locs, labels = plt.xticks();
plt.xticks(locs,map(transform_timestamp,locs));
plt.xlabel(r'Time [HH:MM]',fontsize=14);
plt.ylabel(r'Slope [$^\circ$]',fontsize=14);
plt.legend([r'$\theta_k$'], fontsize=14);
plt.grid(True, linestyle='--')


# Fuel consumption 
plt.subplot(144)
ax1 = plt.gca()
gps_df_flt.loc[st_time:ed_time].plot(y='FUEL_FLOW',ax=ax1);
locs, labels = plt.xticks();
plt.xticks(locs,map(transform_timestamp,locs));
plt.xlabel(r'Time [HH:MM]',fontsize=14);
plt.ylabel(r'Fuel flow [$g/s$]',fontsize=14);
plt.legend([r'$FC_k$'], fontsize=14);
plt.grid(True, linestyle='--')

plt.tight_layout()

plt.savefig('../output/time.pdf',format='pdf', bbox_inches='tight')
```

Resample data, aggregating to reduce the number of data points  

```python
gps_df_agg = gps_df_flt.resample('5Min').mean()
gps_df_agg = gps_df_agg.dropna()
truck_df = gps_df_agg[['SPEED','ACCELERATION','SLOPE','FUEL_FLOW']]
```

Create grid plot 

```python
fig, panels = plt.subplots(nrows=1, ncols=4, figsize = (15,5))
lst_var = ['SPEED','ACCELERATION','SLOPE','FUEL_FLOW']
lst_xlab = [r'$v_k$', r'$a_k$', r'$\theta_k$', r'$FC_k$']
lst_xlim = [(0,100), (-0.1, 0.05), (-0.05, 0.025), (0,1)]
lst_ylim = [(0,17), (0,17), (0,17), (0,1)]
for var, panel, xlab,xlim,ylim in zip(lst_var,panels.flatten(),lst_xlab, lst_xlim, lst_ylim):
    if var == 'FUEL_FLOW':
        panel.hist(gps_df_agg[var], normed=True)
        panel.set_ylabel(r'Freq.', fontsize=16)
    else:
        q = gps_df_agg[var].quantile(0.90)
        df_plot = gps_df_agg[gps_df_agg[var]<q]        
        panel.scatter(df_plot[var],df_plot['FUEL_FLOW'])
        panel.set_xlim(xlim)
        panel.set_ylim(ylim)
        panel.grid(True, linestyle='--')
    panel.set_xlabel(xlab, fontsize=18)
    if var == 'SPEED':
        panel.set_ylabel(r'$FC_k [g/s]$', fontsize=16)
        
        
fig.suptitle(r'Fuel consumption vs. Explanatory variables',fontsize = 20, y = 1.02)        
plt.tight_layout()

plt.savefig('../output/correlation.pdf',format='pdf', bbox_inches='tight')
```
